<h1>Exploratory Data Analysis (EDA)

Imports and Loading Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (10, 6)

df = pd.read_csv("../Dataset/neo_processed.csv", low_memory=False)
```

Sanity Checks

```python
display(df.head())
display(df.info())
display(df.describe(include=[np.number]).T)
```

```python
import os
# Path for saving figures
FIGURE_PATH = "../Figures/"
```

<h2> Physical Parameters

**Distribution (histogram + KDE) of Absolute Magnitude**

H (absolute magnitude) is the brightness measure used as a proxy for size, lower H means brighter/larger. The histogram + KDE reveals the shape of the distribution. We  Look for how hazardous objects compare to the overall H distribution later.

```python
fig, ax = plt.subplots()
sns.histplot(df['H'].dropna(), kde=True, bins=50, ax=ax)
ax.set_title("Distribution of Absolute Magnitude (H)")
ax.set_xlabel("Absolute magnitude (H)")
ax.set_ylabel("Count")
plt.savefig(os.path.join(FIGURE_PATH, 'Distribution of Absolute Magnitude (H).png'))
plt.show()
```

**Distribution of Diameter**

This shows the size distribution (where available). Many diameters are estimated and may have long tails (very large or small). Because diameter is often missing, note the count in describe() from Cell 2. The outliers can be investigated or winsorized for modeling.

```python
fig, ax = plt.subplots()
sns.histplot(df['diameter'].dropna(), bins=60, kde=False, ax=ax)
ax.set_title("Distribution of Diameter (km)")
ax.set_xlabel("Diameter (km)")
ax.set_ylabel("Count")
ax.set_xlim(left=0)
plt.savefig(os.path.join(FIGURE_PATH, 'Distribution of Diameter (km).png'))
plt.show()
```

**Distribution for Albedo**

* This plot shows the reflectivity (albedo) distribution of near-Earth objects.
* Most NEOs have low albedo values (0.05â€“0.3), corresponding to dark, carbon-rich surfaces.
* A few high-albedo outliers may represent metallic or stony asteroids, or they could be due to measurement uncertainty.
* Values above ~0.5 are rare and can be checked as potential outliers or instrument artifacts.
* Because albedo estimates are missing for many entries, this feature will later require imputation or cautious weighting during modeling.

```python
fig, ax = plt.subplots()
sns.histplot(df['albedo'].dropna(), bins=60, kde=False, ax=ax)
ax.set_title("Distribution of Albedo")
ax.set_xlabel("Albedo (reflectivity)")
ax.set_ylabel("Count")
ax.set_xlim(left=0)
plt.savefig(os.path.join(FIGURE_PATH, 'Distribution of Albedo.png'))
plt.show()
```

<h2>Orbital Parameters

**Orbital element distributions (subplots for a, e, i)**

Orbital elements define orbit geometry. a shows orbital sizes (inner vs outer objects), e shows how elliptical orbits are, and i shows tilt. These distributions tell us typical orbit types in our dataset and whether certain orbital ranges correspond to hazardous labels in later modeling.

```python
plt.figure(figsize=(6,4))
sns.histplot(df['a'].dropna(), bins=60)
plt.title("Semi-major axis (a)")
plt.xlabel("a (AU)")
plt.tight_layout()
plt.savefig(os.path.join(FIGURE_PATH, 'Semi-major axis (a).png'))
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['e'].dropna(), bins=60)
plt.title("Eccentricity (e)")
plt.xlabel("e")
plt.tight_layout()
plt.savefig(os.path.join(FIGURE_PATH, 'Ecenntricity (e).png'))
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['i'].dropna(), bins=60)
plt.title("Inclination (i)")
plt.xlabel("i (degrees)")
plt.tight_layout()
plt.savefig(os.path.join(FIGURE_PATH, 'inclination (i).png'))
plt.show()
```

<h2>Relation Between Parameters

**H vs diameter scatter (log scale for diameter)**

H and diameter are inversely related (brighter objects tend to be larger). Using a log y-scale for diameter makes the relationship clearer across many orders of magnitude. We Look for the expected downward trend: lower H â†’ larger diameter. Significant deviation could indicate inconsistent diameter estimates or varying albedos.

```python
fig, ax = plt.subplots()
mask = df['diameter'].notna() & df['H'].notna()
sc = ax.scatter(df.loc[mask, 'H'], df.loc[mask, 'diameter'], alpha=0.6, s=10)
ax.set_yscale('log')
ax.set_xlabel("Absolute magnitude (H)")
ax.set_ylabel("Diameter (km) â€” log scale")
ax.set_title("Scatter: H vs Diameter (log scale)")
plt.savefig(os.path.join(FIGURE_PATH, 'Scatter: H vs Diameter (log scale).png'))
plt.show()
```

**MOID distribution (AU) and histogram with vertical line at 0.05 AU**

MOID is the key orbital proximity metric; NASA uses < 0.05 AU as part of the PHA definition. This histogram shows how many objects cluster near Earth and how many meet that threshold. The vertical line at 0.05 AU visualizes the PHA cutoff and helps you see the potential hazard candidate pool.

```python
fig, ax = plt.subplots()
sns.histplot(df['moid'].dropna(), bins=80, ax=ax)
ax.axvline(0.05, color='red', linestyle='--', linewidth=2)
ax.set_xlim(left=0)
ax.set_title("Distribution of Earth MOID (AU) â€” red dashed line at 0.05 AU (PHA threshold)")
ax.set_xlabel("MOID (AU)")
ax.set_ylabel("Count")
plt.savefig(os.path.join(FIGURE_PATH, 'Distribution of Earth MOID (AU).png'))
plt.show()
```

**Moid vs H scatter colored by hazardous_label**

This scatter helps visualize the two components of the PHA definition together: MOID (x) and H (y). H is inverted because smaller H is brighter/larger. Points colored red are labeled hazardous by the heuristic; we can visually validate the label definition and spot any labeled anomalies (e.g., objects labeled hazardous with moid slightly > 0.05).


```python
fig, ax = plt.subplots()
palette = {0: 'blue', 1: 'red'}
mask = df['moid'].notna() & df['H'].notna()
ax.scatter(df.loc[mask, 'moid'], df.loc[mask, 'H'], c=df.loc[mask, 'hazardous_label'].map(palette), alpha=0.6, s=8)
ax.set_xlabel("MOID (AU)")
ax.set_ylabel("Absolute magnitude (H)")
ax.set_title("MOID vs H â€” colored by hazardous label")
ax.invert_yaxis()
ax.set_xlim(left=0)
plt.savefig(os.path.join(FIGURE_PATH, 'MOID vs H â€” colored by hazardous label.png'))
plt.show()
```

**Correlation heatmap (numeric features)**


The correlation matrix provides a quick view of pairwise linear relationships.

* Strong correlations between H and diameter or between a and per_y.
* Correlations of hazardous_label with moid, H, or derived features (if included). Strong correlations suggest predictive power and inform feature selection for modeling.

```python
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap of Numeric Features")
plt.savefig(os.path.join(FIGURE_PATH, 'Correlation Heatmap of Numeric Features.png'))
plt.show()
```

**Boxplots of moid and H by hazardous_label**

Boxplots compare distributions of moid and H across hazardous vs non-hazardous groups â€” a direct check of whether our label aligns with these key variables. If hazardous objects consistently show lower MOID and lower H, the labeling is consistent. Boxplots also reveal spread and outliers in each group.

```python
fig, axes = plt.subplots(1, 2, figsize=(14,6))
sns.boxplot(x='hazardous_label', y='moid', data=df, ax=axes[0])
axes[0].set_title("MOID by Hazardous Label")
axes[0].set_xlabel("Hazardous label")
axes[0].set_ylabel("MOID (AU)")
sns.boxplot(x='hazardous_label', y='H', data=df, ax=axes[1])
axes[1].set_title("H by Hazardous Label")
axes[1].set_xlabel("Hazardous label")
axes[1].set_ylabel("Absolute magnitude (H)")
axes[1].invert_yaxis()
plt.tight_layout()
plt.savefig(os.path.join(FIGURE_PATH, 'Boxplots of MOID and H by Hazardous Label.png'))
plt.show()
```

**Outlier Detection**
